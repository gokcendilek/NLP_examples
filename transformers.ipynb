```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
```

```python
text="It's great to learn NLP for me"
```

```python
classifier(text)
```

```python
import pandas as pd
outputs=classifier(text)
pd.DataFrame(outputs)
```

```python
classifier=pipeline("zero-shot-classification")
```

```python
text="this is tutorial about Hugging Face"
```

```python
labels=["tech","education","business"]
```

```python
outputs=classifier(text,labels)
pd.DataFrame(outputs)
```

```python
generator=pipeline("text-generation")
```

```python
prompt="this tutorial will walk you through how to"
```

```python
outputs=generator(prompt)
```

```python
print(outputs)
```

```python
outputs=generator(prompt,max_length=50)
print(outputs[0]["generated_text"])
```

```python
generator= pipeline("text-generation", model= "distilgpt2")
```

```python
outputs=generator(prompt,max_length=50)
print(outputs[0]["generated_text"])
```


**SORU-CEVAP YAPMA**


```python
reader=pipeline("question-answering")
```

```python
text="My name is Gökçen Dilek and I love Turkey"
quenstion="What is my name"
```

```python
outputs=reader(question=quenstion, context=text)
pd.DataFrame([outputs])
```


**Özetleme Yapma**


```python
from transformers import pipeline

summarizer = pipeline("summarization")
```

```python
text="""MrBeast veya gerçek ismi ile Jimmy Donaldson (d. 7 Mayıs 1998), Amerikalı bir YouTuber, iş insanı ve hayırseverdir.

Donaldson, 7 Mayıs 1998 doğumludur[1] ve Greenville, Kuzey Karolina'da ikamet etmektedir.[2] 2016 yılında Greenville Hristiyan Akademisi'nden mezun olan Donaldson, Newsweek'e göre YouTuber olarak tam zamanlı bir kariyer yapmak için üniversite eğitimini bırakmıştır.[3] Keemstar ile çevrimiçi bir röportaj sırasında Donaldson, Crohn hastalığı olduğunu belirtmiştir.[4] "MrBro" adlı bir kanala sahip olan CJ Donaldson adında bir ağabeyi vardır.[5]

YouTube kariyeri
Donaldson, YouTube'da video yüklemeye 2012 yılında, 13 yaşındayken "MrBeast6000" adlı kanal ile başlamıştır, 6000 abone Sahibi olunca "MrBeast6000" adındaki şarkısını çıkarmıştır.[6] ancak 2017 yılında yüklediği "100.000'e kadar saymak" konulu videosuna kadar aldığı görüntülenme sayısı düşük kalmış, bu videodan sonra videolarını yükledikçe ise Donaldson'un bilinirliği artmıştır.[7]

Donaldson’un YouTube’da 217 milyondan fazla abonesi ve 38 milyardan fazla görüntülenmesi bulunmaktadır. Dallas merkezli yetenek yönetimi şirketi Night Media tarafından menajerlik görevi üstlenilmektedir.[6]

Donaldson'un yüklediği videolar genel olarak yüksek miktardaki para ve değerli eşyaları evsizler, Twitch yayıncıları veya rastgele kişilere bağışlamasını konu alır.[8] Ayrıca pek çok rekor kırma denemesi ile yarışma düzenlemekte olup, bunların sonucunda kazanan kişilere para ödülü de vermektedir.[9][10] Donaldson ayrıca T-Series ve PewDiePie arasında meydana gelmiş abone sayısı yarışmasında PewDiePie'ı desteklemiş ve bu hususta çeşitli videolar ve yardım kampanyaları yayınlamıştır"""
```

```python
outputs=summarizer(text, max_length=60, clean_up_tokenization_spaces=True)
```

```python
print(outputs[0]["summary_text"])
```
